```python
import matplotlib.pyplot as plt
import numpy as np

amount = [20, 70, 10]
c = plt.cm.BuPu(np.linspace(0, 1, len(amount)))
l = ["сон", "учёба", "еда"]
plt.pie(amount, labels=l, colors=c, autopct='%1.f%%')
plt.title("Распредление моего времени в течении дня")
plt.show()
```

```python
x = ['Июль', 'Август', 'Сентябрь', 'Октябрь']
y = [31, 21, 0, 0]

plt.bar(x, y, label='Данные с календаря', alpha=0.7 )
plt.plot(x, y, color='red', markersize=7)
plt.xlabel('Месяц года')
plt.ylabel('% от общего времени')
plt.title('Количество свободного времени :(')
plt.legend(loc='upper right')
plt.show()
```

```python
x = np.linspace(0, 100, 10)
y = np.sqrt(x)
plt.scatter(x, y, c=x, cmap='gnuplot_r')
plt.xlabel('Ось х') 
plt.ylabel('Ось y') 
plt.title('Огонёк') 
plt.grid(True)  
plt.show()
```

```python
def simple_mult(x):
    mult = []
    for i in range(1, x+1):
        if x % i == 0:
            mult.append(i)
    return mult
print(f'Делители: {simple_mult(int(input('Введите число:')))}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def MNK(x, y):
    a1 = a2 = a3 = a4 = a5 = j = 0  
    for i in range(0, len(x)):
        a1 += x[i] * y[i]
        a2 += x[i]
        a3 += y[i]
        a4 += x[i] ** 2
        j += 1
    a5 = a2**2
    a = (len(x) * a1 - a2 * a3)/(len(x) * a4 - a5)
    b = 1/len(x)*(a3 - a * a2)
    return [a, b]

x, y = [1,2,3,4] , [1,4,9,16]
line = MNK(x,y)
plt.plot(x, y, 'ro')
x2 = np.arange( min(x)*0.85, max(x)*1.25)
y2 = x2 * line[0] + line[1]
plt.plot(x2, y2)
plt.plot
plt.show
```
